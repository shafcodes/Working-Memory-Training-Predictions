## Logistic Regression Model

```python
def get_train_test_data(df, upto, demographics, all_demographics, time):
    ''' Returns training (0.8) and testing data (0.2) with random state specified as 42.
        upto: returns score data from session 1 upto session x where x = upto
        gender: adds gender binary indicator to training and testing data
        age: adds age factor to training and testing data
        diff_time: adds difference in time between current session and last session
        avg_time: adds the average time between training sessions in the study
        NOTE: Change demographic names when analyzing different datasets (lab data vs lumosity data)
    '''
    # Drop NA cols for training
    consider_drop_cols = ["Age_y", "Gender_y", "compensation", "domestic", "supervised", "binary_growth"] + ["Training " + str(i) + " Perf_y" for i in range(1, 15+1)] + ["Training " + str(i) + " Perf_y" for i in range(1, upto+1)] + ["AVG Time to " + str(upto-1)]
    trimmed_df = df.dropna(subset=consider_drop_cols)

    # Gather training factors (X)
    if not demographics and not all_demographics and not time:  # trains only on scores
        X = trimmed_df[["Training " + str(i) + " Perf_y" for i in range(1, upto+1)]]
    elif demographics and not all_demographics and not time:  # trains on scores + demographics
        X = trimmed_df[["Training " + str(i) + " Perf_y" for i in range(1, upto+1)] + ["Age_y", "Gender_y"]]
    elif demographics and all_demographics and not time:  # trains on scores + all demographics
        X = trimmed_df[["Training " + str(i) + " Perf_y" for i in range(1, upto+1)] + ["Age_y", "Gender_y", "supervised", "compensation", "domestic"]]  # add "binary_growth" here
    elif demographics and not all_demographics and time:  # trains on scores + demographics + time
        X = trimmed_df[["Time Elapsed " + str(upto-1)] + (["Training " + str(i) + " Perf_y" for i in range(1, upto+1)]) + ["AVG Time to " + str(upto-1)] + ["Age_y", "Gender_y", "supervised", "compensation", "domestic"]]
    else:
        print("Incorrect model training specified.")
    
    Y = trimmed_df["resp"]
    
    return train_test_split(X, Y, test_size = 0.2, random_state=42)
```

```python
def Logit(X_train, X_test, y_train, y_test):
    ''' Train input data with logistic regression model using sklearn.
        Returns predicted y training and predicted y testing along with their actual y train and tests.
    '''
    # Train logistic regression model
    log = LogisticRegression(solver='lbfgs', max_iter=800)
    logm = log.fit(X_train, y_train)
    
    # Predict based on model
    y_test_pred = logm.predict(X_test)   # TESTING DATA
    y_train_pred = logm.predict(X_train) # TRAINING DATA
    return y_train, y_train_pred, y_test, y_test_pred
```

```python
def record_metrics(y_train, y_train_pred, y_test, y_test_pred, metrics_list):
    ''' Returns recorded training and testing metrics from testing the model:
        Structure: 1d array
                    - list of all metrics for that session
                    - example: (session1) [accuracy, auc, r^2]
    '''
    all_train_metrics = []
    all_test_metrics = []
    
    for metric in metrics_list:
        all_train_metrics.append(metric(y_train, y_train_pred))  # add metric
        all_test_metrics.append(metric(y_test, y_test_pred))
    
    return all_test_metrics #, all_train_metrics <- don't care about training metrics
```

```python
def generate_dummy_col():
    consider_drop_cols = ["Age_y", "Gender_y"] + ["Training " + str(i) + " Perf_y" for i in range(1, 15+1)]
    trimmed_df = df.dropna(subset=consider_drop_cols)
    initial_mean = trimmed_df["Training 1 Perf_y"].mean()
    initial_std = trimmed_df["Training 1 Perf_y"].std()
    intial_size = trimmed_df["Training 1 Perf_y"].shape[0]
    new_intial = np.random.normal(initial_mean, initial_std, intial_size)
    df_with_dummy = trimmed_df.copy()
    df_with_dummy.insert(loc=2, column='-1', value=new_intial)
    return df_with_dummy
```

```python
dummy_baselines = []
for i in range(100):
    df_with_dummy = generate_dummy_col()  # new dataframe with dummy variable added

    X_dummy = df_with_dummy[["-1", "Age_y", "Gender_y"]]
    Y_dummy = df_with_dummy["resp"] 
    X_dummy_train, X_dummy_test, y_dummy_train, y_dummy_test = train_test_split(X_dummy, Y_dummy, test_size = 0.2)
    dummy_baseline = record_metrics(*Logit(X_dummy_train, X_dummy_test, y_dummy_train, y_dummy_test), [metrics.accuracy_score])
    dummy_baselines.append(dummy_baseline)
    
# Calculating average dummy_baseline
total = 0
for dummy_baseline in dummy_baselines:
    total += dummy_baseline[0]
avg_dummy_baseline = total / len(dummy_baselines)
print(avg_dummy_baseline)
```

```python
def get_logit_metrics():
    metrics_list = [metrics.accuracy_score, metrics.precision_score, metrics.recall_score, metrics.f1_score, metrics.roc_auc_score]
    all_only_scores_metrics = []
    all_scores_dems_metrics = []
    all_scores_alldems_metrics = []
    all_scores_dems_times_metrics = []
    
    for i in range(1, 16):
        only_scores_metrics = record_metrics(*Logit(*get_train_test_data(df, i, False, False, False)), metrics_list) # only scores
        scores_dems_metrics = record_metrics(*Logit(*get_train_test_data(df, i, True, False, False)), metrics_list) # scores + dems
        scores_alldems_metrics = record_metrics(*Logit(*get_train_test_data(df, i, True, True, False)), metrics_list) # scores + dems + all_dems
        scores_dems_times_metrics = record_metrics(*Logit(*get_train_test_data(df, i, True, False, True)), metrics_list) # scores + dems + time
        all_only_scores_metrics.append(only_scores_metrics)
        all_scores_dems_metrics.append(scores_dems_metrics)
        all_scores_alldems_metrics.append(scores_alldems_metrics)
        all_scores_dems_times_metrics.append(scores_dems_times_metrics)
    
    return all_only_scores_metrics, all_scores_dems_metrics, all_scores_alldems_metrics, all_scores_dems_times_metrics
```

## Random Forest

```python
# Default hyperparameters are depth=10 and number of estimators = 20
def Forest(X_train, X_test, y_train, y_test, depth=10, estimators=20):
    ''' Train input data with a random forest classification model using sklearn.
        Returns predicted y training and predicted y testing along with their actual y train and tests.
    '''
    # Train random forest classifier model
    forest = RandomForestClassifier(n_estimators=estimators, max_depth=depth)
    mod = forest.fit(X_train, y_train)
    
    # Predict based on model
    y_test_pred = mod.predict(X_test)   # TESTING DATA
    y_train_pred = mod.predict(X_train) # TRAINING DATA
    return y_train, y_train_pred, y_test, y_test_pred
```

```python
def get_forest_metrics():
    metrics_list = [metrics.accuracy_score]
    all_only_scores_metrics = []
    all_scores_dems_metrics = []
    all_scores_alldems_metrics = []
    all_scores_dems_times_metrics = []
    
    for i in range(1, 16):
        only_scores_metrics = record_metrics(*Forest(*get_train_test_data(df, i, False, False, False)), metrics_list) # only scores
        scores_dems_metrics = record_metrics(*Forest(*get_train_test_data(df, i, True, False, False)), metrics_list) # scores + dems
        scores_alldems_metrics = record_metrics(*Forest(*get_train_test_data(df, i, True, True, False)), metrics_list) # scores + dems + all_dems
        scores_dems_times_metrics = record_metrics(*Forest(*get_train_test_data(df, i, True, False, True)), metrics_list) # scores + dems + time
        all_only_scores_metrics.append(only_scores_metrics)
        all_scores_dems_metrics.append(scores_dems_metrics)
        all_scores_alldems_metrics.append(scores_alldems_metrics)
        all_scores_dems_times_metrics.append(scores_dems_times_metrics)
    
    return all_only_scores_metrics, all_scores_dems_metrics, all_scores_alldems_metrics, all_scores_dems_times_metrics
```

```python
# getting dummy metric for random forest classifier
dummy_baselines = []
for i in range(100):
    df_with_dummy = generate_dummy_col()  # new dataframe with dummy variable added

    X_dummy = df_with_dummy[["-1", "Age_y", "Gender_y"]]
    Y_dummy = df_with_dummy["resp"] 
    X_dummy_train, X_dummy_test, y_dummy_train, y_dummy_test = train_test_split(X_dummy, Y_dummy, test_size = 0.2)
    dummy_baseline = record_metrics(*Forest(X_dummy_train, X_dummy_test, y_dummy_train, y_dummy_test), [metrics.accuracy_score])
    dummy_baselines.append(dummy_baseline)
    
# Calculating average dummy_baseline
total = 0
for dummy_baseline in dummy_baselines:
    total += dummy_baseline[0]
avg_forest_dummy_baseline = total / len(dummy_baselines)
print(avg_forest_dummy_baseline)
```

```python
def transform_metrics(metric_scores):
    metric_scores_list = []
    for score in metric_scores:
        metric_scores_list.append(score)
    return metric_scores_list    
```

## Optimizing Forest Model

```python
def get_session_forest_metric(sessionid):
    metrics_list = [metrics.accuracy_score]
    session_only_scores_metrics = []
    session_scores_dems_metrics = []
    session_scores_alldems_metrics = []
    session_scores_dems_times_metrics = []
    
    for n_estimators in range(1, 50):
        only_scores_metrics = record_metrics(*Forest(*get_train_test_data(df, sessionid, False, False, False), 10, n_estimators), metrics_list) # only scores
        scores_dems_metrics = record_metrics(*Forest(*get_train_test_data(df, sessionid, True, False, False), 10, n_estimators), metrics_list) # scores + dems
        scores_alldems_metrics = record_metrics(*Forest(*get_train_test_data(df, sessionid, True, True, False), 10, n_estimators), metrics_list) # scores + dems + all_dems
        scores_dems_times_metrics = record_metrics(*Forest(*get_train_test_data(df, sessionid, True, False, True), 10, n_estimators), metrics_list) # scores + dems + time
        session_only_scores_metrics.append(only_scores_metrics)
        session_scores_dems_metrics.append(scores_dems_metrics)
        session_scores_alldems_metrics.append(scores_alldems_metrics)
        session_scores_dems_times_metrics.append(scores_dems_times_metrics)
    
    return session_only_scores_metrics, session_scores_dems_metrics, session_scores_alldems_metrics, session_scores_dems_times_metrics
```
