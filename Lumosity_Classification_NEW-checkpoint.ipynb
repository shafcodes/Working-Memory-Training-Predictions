```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandasql as ps
from scipy import stats
from sklearn import tree
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn import metrics, linear_model, neighbors
from statistics import mean
import matplotlib.font_manager as font_manager
np.random.seed(0)
```

```python
def dTree(X_train, X_test, y_train, y_test, scoring, d):
    dt = tree.DecisionTreeClassifier(max_depth = d)
    dt = dt.fit(X_train, y_train)
    y_test_pred = dt.predict(X_test)
    y_train_pred = dt.predict(X_train)
    return [scoring(y_train, y_train_pred), scoring(y_test, y_test_pred)]
```

```python
def Logit(X_train, X_test, y_train, y_test, scoring):
    log = LogisticRegression(solver='lbfgs', max_iter=200)
    logm = log.fit(X_train, y_train)
    y_test_pred = logm.predict(X_test)
    y_train_pred = logm.predict(X_train)
    return [scoring(y_train, y_train_pred), scoring(y_test, y_test_pred)]
```

```python
def KNN(X_train, X_test, y_train, y_test, scoring, n):
    knn = KNeighborsClassifier(n_neighbors=n)
    knnm = knn.fit(X_train, y_train)
    y_test_pred = knnm.predict(X_test)
    y_train_pred = knnm.predict(X_train)
    return [scoring(y_train, y_train_pred), scoring(y_test, y_test_pred)]
```

```python
def Forest(X_train, X_test, y_train, y_test, scoring, t, d):
    forest = RandomForestClassifier(n_estimators=t,max_depth=d)
    forestm = forest.fit(X_train, y_train)
    y_test_pred = forestm.predict(X_test)
    y_train_pred = forestm.predict(X_train)
    return [scoring(y_train, y_train_pred), scoring(y_test, y_test_pred)]
```

```python
def SVM(X_train, X_test, y_train, y_test, scoring):
    svmo = svm.LinearSVC(dual=False)
    svmm = svmo.fit(X_train, y_train)
    y_test_pred = svmm.predict(X_test)
    y_train_pred = svmm.predict(X_train)
    return [scoring(y_train, y_train_pred), scoring(y_test, y_test_pred)]
```

```python
#TT for all, TF for edu, FT for age, FF for none
def train_test(df, Upto, demo = False, age = False):
    if demo and age:
        X = df[[str(i) for i in range(Upto)] + ["age", "edubin"]]
    elif not demo and not age:
        X = df[[str(i) for i in range(Upto)]]
    elif not demo and age:
        X = df[[str(i) for i in range(Upto)] + ["age"]]
    elif demo and not age:
        X = df[[str(i) for i in range(Upto)] + ["edubin"]]
    Y = df["perf"] 
    return train_test_split(X, Y, test_size = 0.2, random_state=42)
```

```python
def generate_dummy_col():
    initial_mean = df["0"].mean()
    initial_std = df["0"].std()
    intial_size = df["0"].shape[0]
    new_intial = np.random.normal(initial_mean, initial_std, intial_size)
    df_dum = df.copy()
    df_dum.insert(loc=2, column='-1', value=new_intial)
    return df_dum
```
